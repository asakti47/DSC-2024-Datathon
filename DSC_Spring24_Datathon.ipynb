# **Analyze Similarity/Fit between Displaced and Viable To-Move Areas**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import euclidean

# Euclidean distance function for latitude, longitude
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Normalize economic data only
scaler = MinMaxScaler()
economic_columns = ['families_median_income', 'BUILDVALUE', 'median_monthly_costs_occupied_units']

# Copy DataFrames to new DataFrames to save normalized values
displaced_df_normal = displaced_df.copy()
moving_df_normal = moving_df.copy()
social_connect_normal = social_connect_filtered.copy()

# Apply normalization
displaced_df_normal[economic_columns] = scaler.fit_transform(displaced_df_normal[economic_columns])
moving_df_normal[economic_columns] = scaler.fit_transform(moving_df_normal[economic_columns])
social_connect_normal['scaled_sci'] = scaler.fit_transform(social_connect_normal[['scaled_sci']])

# Calculate the maximum geographic distance for scaling
max_geo_distance = 0
for _, row1 in displaced_df.iterrows():
    for _, row2 in moving_df.iterrows():
        dist = euclidean_distance(row1['longitude'], row1['latitude'],
                                  row2['longitude'], row2['latitude'])
        if dist > max_geo_distance:
            max_geo_distance = dist

# Weights
w1, w2, w3, w4, w5 = 1, 1, 1, 2, 5  # Adjust w5 based on experimentation

# Calculate composite fit/similarity score
results = []
for index_displaced, row_displaced in displaced_df_normal.iterrows():
    for index_moving, row_moving in moving_df_normal.iterrows():
        # Calculate each distance for economic factors
        D1 = euclidean(row_displaced[['families_median_income']], row_moving[['families_median_income']])
        D2 = euclidean(row_displaced[['BUILDVALUE']], row_moving[['BUILDVALUE']])
        D3 = euclidean(row_displaced[['median_monthly_costs_occupied_units']], row_moving[['median_monthly_costs_occupied_units']])

        # Calculate geographic distance and scale it
        geo_distance = euclidean_distance(displaced_df['longitude'][index_displaced], displaced_df['latitude'][index_displaced],
                                          moving_df['longitude'][index_moving], moving_df['latitude'][index_moving]) / max_geo_distance

        # Fetch the precomputed sci_score
        sci_score = social_connect_normal.loc[
            (social_connect_normal['user_loc'] == row_displaced['STCOFIPS']) &
            (social_connect_normal['fr_loc'] == row_moving['STCOFIPS']),
            'scaled_sci'
        ].values
        if len(sci_score) == 0:
            sci_score = [0]  # Default low connection score

        # Combine the distances with weights
        combined_score = (w1 * D1) + (w2 * D2) + (w3 * D3) + (w4 * sci_score[0]) + (w5 * geo_distance)

        results.append({
            'Displaced_TRACTFIPS': row_displaced['TRACTFIPS'],
            'Moving_TRACTFIPS': row_moving['TRACTFIPS'],
            'Combined_Score': combined_score,
            'w1 * D1': (w1) * D1,
            'w2 * D2': (w2) * D2,
            'w3 * D3': (w3) * D3,
            'w4 * D4': (w4) * sci_score[0],
            'w5 * D5': (w5) * geo_distance
        })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Normalize the composite score

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the 'Combined_Score' column
results_df_normal = results_df.copy()
results_df_normal['Combined_Score_MinMax'] = scaler.fit_transform(results_df_normal['Combined_Score'].values.reshape(-1, 1))
results_df_normal
```

```python
# Create an index score
# Multiplied by 98, because index score should logically not be perfect due to challenges associated with displacement
results_df_normal['index_score'] = ((1 - results_df_normal['Combined_Score_MinMax']) * 98) + 1
results_df_normal
```

```python
# First, group the DataFrame by 'Displaced_TRACTFIPS'
grouped = results_df_normal.groupby('Displaced_TRACTFIPS')

# Now iterate over each group, calculate the top 5 minimum 'Combined_Score', and display them
for tractfips, group in grouped:
    # Get the top 5 entries with the smallest 'Combined_Score' within this group
    top_five = group.nsmallest(5, 'Combined_Score')
    print(f"Top 5 relocation options for Displaced TRACTFIPS {tractfips}:")
    display(top_five)  # Using display to show the DataFrame in a more formatted style
    print("\n")  # Adds a newline for better readability between tables
```
